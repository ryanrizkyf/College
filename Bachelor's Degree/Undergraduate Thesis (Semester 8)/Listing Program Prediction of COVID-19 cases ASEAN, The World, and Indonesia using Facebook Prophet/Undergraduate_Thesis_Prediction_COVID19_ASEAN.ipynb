# Preprocessing Data

```python
def get_json(api_url):
	response = requests.get(api_url)
	if response.status_code == 200:
		return json.loads(response.content.decode('utf-8'))
	else:
		return None
```

```python
url_sg = 'https://api.covid19api.com/dayone/country/singapore'

url_my = 'https://api.covid19api.com/dayone/country/malaysia'

url_th = 'https://api.covid19api.com/dayone/country/thailand'

url_vn = 'https://api.covid19api.com/dayone/country/vietnam'

url_kh = 'https://api.covid19api.com/dayone/country/cambodia'

url_mm = 'https://api.covid19api.com/dayone/country/myanmar'

url_ph = 'https://api.covid19api.com/dayone/country/philippines'

url_bn = 'https://api.covid19api.com/dayone/country/brunei'

url_la = 'https://api.covid19api.com/dayone/country/laos'
```

# Prediksi Kasus di Singapore

```python
from fbprophet import Prophet
```

```python
df_conf2 = df_conf_sg.copy()
```

```python
df_conf2.columns
```

```python
df_conf2.columns=['ds','y']
```

```python
model_sg = Prophet(interval_width=0.95)
model_sg.fit(df_conf2)
```

```python
future_sg = model_sg.make_future_dataframe(periods=60)
```

```python
future_sg.tail()
```

```python
forecast_conf_sg = model_sg.predict(future_sg)
```

```python
pd.options.display.float_format = '{: .2f}'.format
```

```python
forecast_conf_sg.tail()
```

```python
forecast_conf_sg[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
forecast_conf_plot_sg = model_sg.plot(forecast_conf_sg, figsize=(16, 6), xlabel='', ylabel='Prediksi Kasus')
ax = forecast_conf_plot_sg.gca()
ax.set_title('Prediksi Total Kasus Positif COVID-19 di Singapore', size=22)
```

```python
conf_forecast_plot_component_sg = model_sg.plot_components(forecast_conf_sg, figsize=(16, 6))
```

# Cross-Validation the Time Series Model using Prophet - Singapore

```python
from fbprophet.plot import add_changepoints_to_plot
```

```python
fig=model_sg.plot(forecast_conf_sg, figsize=(16, 6))

a=add_changepoints_to_plot(fig.gca(), model_sg, forecast_conf_sg)
```

```python
from fbprophet.diagnostics import cross_validation
```

```python
df_cv_sg=cross_validation(model_sg, horizon='60 days', period='30 days', initial='90 days')
```

```python
df_cv_sg.head()
```

```python
df_cv_sg.shape
```

# Obtain Performance Metrics for the Data - Singapore

```python
from fbprophet.diagnostics import performance_metrics
```

```python
df_performance_sg=performance_metrics(df_cv_sg)
```

```python
df_performance_sg.head()
```

```python
df_performance_sg.tail()
```

```python
from fbprophet.plot import plot_cross_validation_metric
```

```python
df_performance_sg=plot_cross_validation_metric(df_cv_sg, metric='mape', figsize=(16, 6))
```
